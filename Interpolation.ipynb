2. Proof: independent of basis it's always the same polynomial

The polynomial will always be same regardless of what basis we use because the interpolating polynomial is always unique. 

Proof by contradiction:

If x0, x1, ...xn are distinct points then for the values of y0, y1, ...yn, there exist a polynomial $p_n(x_i) = y_i$ for $0 <= i <= n$ n+1 points. Assume interpolating polynomials are not unique and there exist two interpolating polynomials $p_n$ and $q_n$ of degree $<= n$. 

Then,
$$p_n(x_i) = y_i$$

$$q_n(x_i) = y_i$$ 

$$(p_n - q_n)(x_i) = 0$$

The degree of $p_n - q_n <= n$. A polynomial of degree $n$ has at most $n$ zeroes unless it is the zero polynomial.

If $p_n \neq q_n$, then, we have $n+1$ zeroes, which contradicts the n zeroes of n polynomial property.

So therefore $p_n - q_n = 0$ and this shows that $p_n = q_n$. This proves that a polynomial will always be same regardless of what basis we use because the interpolating polynomial is always unique.

````python
# Compute the Lagrange basis (\ell_i(x))
def lagrange_basis(x, data):
    """Compute Lagrange basis at x given N data points
    
    params:
    -------
        x: ndarray
            1-d Array of floats
        data: ndarray of shape (N,2)
            2-d Array of data where each row is [ x_i, y_i ]
    returns:
    --------
        basis: ndarray of shape (N, x.shape)
            : 2-D array of lagrange basis functions evaluated at x
            
    """
    basis = numpy.ones((data.shape[0], x.shape[0]))
    for i in range(data.shape[0]):
        for j in range(data.shape[0]):
            if i != j:
                basis[i, :] *= (x - data[j, 0]) / (data[i, 0] - data[j, 0])
    return basis

def poly_interpolant(x, data):
    """Compute polynomial interpolant of (x,y) using Lagrange basis"""
    P = numpy.zeros(x.shape[0])
    basis = lagrange_basis(x, data)
    for n in range(data.shape[0]):
        P += basis[n, :] * data[n, 1]
    return P

def gaussian_fn(x):
    return numpy.exp(-x**2)
````

Given this matrix $D$, it is easy to test it for various functions and meshes

````python
from  fdcoeffV import fdcoeffV

def D1(x):
    """
    Write a good doc-string
    
    x: ndarray of coordinates
    D: ndarray
        Matrix (2-D ndarray ) to calculate consistent 2nd order first derivative
    """
    # YOUR CODE HERE
    n = 2
    k = 1
    scale = (x[1]-x[0])**k
    matrix_size = x.shape[0]
    A = numpy.zeros((matrix_size, matrix_size))
    coeff = fdcoeffV(k,  x[0],x[:3])*scale
    A[0, 0] = 1
    A[0, 1] = coeff[1]
    A[0, 2] = coeff[2]
    coeff = fdcoeffV(k,x[-1],x[-3:])*scale
    A[matrix_size-1, 0] = 1
    A[matrix_size-1, 1] = coeff[1]
    A[matrix_size-1, 2] = coeff[0]
    for i in range(1, matrix_size-1):
        coeff = fdcoeffV(k,  x[i], x[i-1:i+2])*scale
        A[i, i-1] = coeff[0]
        A[i, i] = coeff[1]
        A[i, i+1] = coeff[2]
    return A
    
````
